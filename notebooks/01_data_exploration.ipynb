## üìñ About the Dataset
This dataset contains features computed from digitized images of **fine needle aspirates (FNA)** of breast masses.  
These features describe the characteristics of the **cell nuclei** present in the image.

The data was originally published in:  
*K. P. Bennett and O. L. Mangasarian (1992): "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 23‚Äì34.*

This database is available on the [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic))  
and also mirrored on the UW CS FTP server:ftp ftp.cs.wisc.edu

## üß† Attribute Information
1. **ID number**  
2. **Diagnosis:**  
   - M = Malignant (cancerous)  
   - B = Benign (non-cancerous)  
3. **Features (3‚Äì32):**  
   Ten real-valued features are computed for each cell nucleus:

   - **Radius:** Mean of distances from center to points on the perimeter  
   - **Texture:** Standard deviation of gray-scale values  
   - **Perimeter**  
   - **Area**  
   - **Smoothness:** Local variation in radius lengths  
   - **Compactness:** (perimeter¬≤ / area ‚àí 1.0)  
   - **Concavity:** Severity of concave portions of the contour  
   - **Concave points:** Number of concave portions of the contour  
   - **Symmetry**  
   - **Fractal dimension:** "Coastline approximation" ‚àí 1

For each feature, **three metrics** were calculated:
- Mean value  
- Standard error (SE)  
- ‚ÄúWorst‚Äù or largest (mean of the three largest values)

In total, this results in **30 features** per sample (from columns 3‚Äì32).

---

## üìä Dataset Details
- **Number of instances:** 569  
- **Number of features:** 32 (including ID and Diagnosis)  
- **Feature values:** Recorded to four significant digits  
- **Missing values:** None  
- **Class distribution:**
  - 357 Benign  
  - 212 Malignant  

---

## üéØ Objective
The purpose of this dataset is to develop and evaluate models that can classify breast tumors as **benign or malignant**  
based on quantitative features computed from digitized images.
